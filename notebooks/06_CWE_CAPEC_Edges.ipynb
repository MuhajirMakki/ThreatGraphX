### BUILD CAPEC HIERARCHY EDGES

```python
for attack_pattern in root.iter():

    if "Attack_Pattern" in attack_pattern.tag and "ID" in attack_pattern.attrib:

        print("\nCAPEC:", attack_pattern.attrib)

        print("CHILDREN TAGS:")

        for child in list(attack_pattern)[:20]:
            print(child.tag)

        break
```

```python
relation_hits = []

for elem in root.iter():

    tag = elem.tag.split("}")[-1]

    if "ChildOf" in tag or "CanPrecede" in tag or "PeerOf" in tag:
        relation_hits.append(tag)

len(relation_hits), relation_hits[:20]
```

```python
for elem in root.iter():

    if "CAPEC_ID" in elem.attrib:

        print(elem.tag, elem.attrib)

        break
```

###  Extract Relationships

```python
edges = []

for attack_pattern in root.iter():

    if "Attack_Pattern" in attack_pattern.tag and "ID" in attack_pattern.attrib:

        source_id = "CAPEC-" + attack_pattern.attrib["ID"]

        for child in attack_pattern:

            child_tag = child.tag.split("}")[-1]

            if child_tag == "Related_Attack_Patterns":

                for rel in child:

                    rel_tag = rel.tag.split("}")[-1]

                    if rel_tag in ["Related_Attack_Pattern"]:

                        target = rel.attrib.get("CAPEC_ID")

                        nature = rel.attrib.get("Nature")

                        if target and nature in ["ChildOf", "CanPrecede", "PeerOf"]:

                            edges.append([
                                source_id,
                                nature,
                                "CAPEC-" + target
                            ])
```

### BUILD CWE → CAPEC EDGES

```python
edges = []
```

### Extract CWE → CAPEC relationships

```python
for attack_pattern in root.iter():

    if "Attack_Pattern" in attack_pattern.tag and "ID" in attack_pattern.attrib:

        capec_id = "CAPEC-" + attack_pattern.attrib["ID"]

        for child in attack_pattern:

            child_tag = child.tag.split("}")[-1]

            if child_tag == "Related_Weaknesses":

                for weakness in child:

                    cwe_id = weakness.attrib.get("CWE_ID")

                    if cwe_id:

                        edges.append([
                            "CWE-" + cwe_id,
                            "mapped_to",
                            capec_id
                        ])
```

### Convert to DataFrame

```python
import pandas as pd

cwe_capec_df = pd.DataFrame(
    edges,
    columns=["source", "relation", "target"]
)

cwe_capec_df.head()
```

### Remove duplicates

```python
cwe_capec_df = cwe_capec_df.drop_duplicates()

len(cwe_capec_df)
```
