### Load CVE Data

```python
import json
from pathlib import Path

project = Path(r"C:\Users\HP\Documents\python\bio informatics\final exam\ThreatGraphx")

cve_file = project / "data" / "raw" / "cve" / "nvdcve-2.0-recent.json"

with open(cve_file, "r", encoding="utf-8") as f:
    cve_data = json.load(f)
```

### Edge List

```python
edges = []
```

### Extract Relationships

```python
for item in cve_data["vulnerabilities"]:
    
    cve_id = item["cve"]["id"]
    
    weaknesses = item["cve"].get("weaknesses", [])
    
    for weakness in weaknesses:
        
        descriptions = weakness.get("description", [])
        
        for desc in descriptions:
            
            value = desc.get("value", "")
            
            if value.startswith("CWE-"):
                
                edges.append(
                    [
                        cve_id,
                        "has_weakness",
                        value
                    ]
                )
```

### make DataFrame

```python
import pandas as pd

edge_df = pd.DataFrame(
    edges,
    columns=[
        "source",
        "relation",
        "target"
    ]
)

edge_df.head()
```

### Remove Duplicates

```python
edge_df = edge_df.drop_duplicates()

len(edge_df)
```

### Save

```python
output_file = project / "data" / "processed" / "cve_cwe_edges.csv"

edge_df.to_csv(output_file, index=False)

print("saved")
```
