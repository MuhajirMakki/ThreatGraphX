### Load CWE File

```python
import xml.etree.ElementTree as ET
from pathlib import Path

project = Path(r"C:\Users\HP\Documents\python\bio informatics\final exam\ThreatGraphx")

cwe_file = project / "data" / "raw" / "cwe" / "1435.xml"

tree = ET.parse(cwe_file)
root = tree.getroot()

print(root.tag)
```

### Root Information

```python
print(root.attrib)
```

### Weakness Elements

```python
for elem in root.iter():
    if "Weakness" in elem.tag:
        print(elem.tag)
        break
```

### Exploring One Weakness

```python
for elem in root.iter():
    if elem.tag.endswith("Weakness"):
        print(elem.attrib)
        break
```

### Extract IDs

```python
cwe_ids = []

for elem in root.iter():
    if elem.tag.endswith("Weakness"):
        
        if "ID" in elem.attrib:
            cwe_ids.append(
                "CWE-" + elem.attrib["ID"]
            )

len(cwe_ids)
```

### checking first 10

```python
cwe_ids[:10]
```

### creating dataframe

```python
import pandas as pd

cwe_df = pd.DataFrame({
    "cwe_id": cwe_ids
})

cwe_df.head()
```

### Save

```python
output_file = project / "data" / "processed" / "cwe_nodes.csv"

cwe_df.to_csv(output_file, index=False)

print("saved")
```
