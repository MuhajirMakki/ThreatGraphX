```python
import pandas as pd
import numpy as np
import pickle

from pathlib import Path
```

```python
project = Path(
    r"C:\Users\HP\Documents\python\bio informatics\final exam\ThreatGraphx"
)
```

### Load All Edge Files

```python
cve_cwe = pd.read_csv(
    project / "data" / "processed" / "cve_cwe_edges.csv"
)

cwe_capec = pd.read_csv(
    project / "data" / "processed" / "cwe_capec_edges.csv"
)

capec_attack = pd.read_csv(
    project / "data" / "processed" / "capec_attack_edges.csv"
)

attack_mitigation = pd.read_csv(
    project / "data" / "processed" / "attack_mitigation_edges.csv"
)

software_edges = pd.read_csv(
    project / "data" / "processed" / "software_edges.csv"
)
```

### Merge

```python
edges = pd.concat(
    [
        cve_cwe,
        cwe_capec,
        capec_attack,
        attack_mitigation,
        software_edges
    ],
    ignore_index=True
)

print(len(edges))
```

### Load Dictionaries

```python
with open(
    project / "data" / "ml_ready" / "node2id.pkl",
    "rb"
) as f:

    node2id = pickle.load(f)

with open(
    project / "data" / "ml_ready" / "relation2id.pkl",
    "rb"
) as f:

    relation2id = pickle.load(f)
```

### Create edge_index

```python
edge_index = np.array([
    
    [
        node2id[src]
        for src in edges["source"]
    ],

    [
        node2id[dst]
        for dst in edges["target"]
    ]

])

print(edge_index.shape)
```

### Create edge_type

```python
edge_type = np.array([
    
    relation2id[r]
    
    for r in edges["relation"]
])

print(edge_type.shape)
```

### Save

```python
ml_dir = project / "data" / "ml_ready"

np.save(
    ml_dir / "edge_index.npy",
    edge_index
)

np.save(
    ml_dir / "edge_type.npy",
    edge_type
)

print("saved")
```
