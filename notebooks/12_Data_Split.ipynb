### imports

```python
import pandas as pd
import numpy as np

from pathlib import Path

from sklearn.model_selection import train_test_split
```

### Project Path

```python
project = Path(
    r"C:\Users\HP\Documents\python\bio informatics\final exam\ThreatGraphx"
)
```

### Load Unified Edge File

```python
cve_cwe = pd.read_csv(
    project / "data" / "processed" / "cve_cwe_edges.csv"
)

cwe_capec = pd.read_csv(
    project / "data" / "processed" / "cwe_capec_edges.csv"
)

capec_attack = pd.read_csv(
    project / "data" / "processed" / "capec_attack_edges.csv"
)

attack_mitigation = pd.read_csv(
    project / "data" / "processed" / "attack_mitigation_edges.csv"
)

software_edges = pd.read_csv(
    project / "data" / "processed" / "software_edges.csv"
)

edges = pd.concat(
    [
        cve_cwe,
        cwe_capec,
        capec_attack,
        attack_mitigation,
        software_edges
    ],
    ignore_index=True
)

print(len(edges))
```

### Train Split

```python
train_edges, temp_edges = train_test_split(
    edges,
    test_size=0.30,
    random_state=42,
    shuffle=True
)
```

### Validation + Test Split

```python
val_edges, test_edges = train_test_split(
    temp_edges,
    test_size=0.50,
    random_state=42,
    shuffle=True
)
```

### Verify Counts

```python
print("Train:", len(train_edges))
print("Validation:", len(val_edges))
print("Test:", len(test_edges))
```

### Check Ratios

```python
total = len(edges)

print("Train %:", round(len(train_edges)/total*100, 2))
print("Val %:", round(len(val_edges)/total*100, 2))
print("Test %:", round(len(test_edges)/total*100, 2))
```

### Save Split Files

```python
split_dir = (
    project
    / "data"
    / "splits"
)

split_dir.mkdir(exist_ok=True)
```

### Save Train

```python
train_edges.to_csv(
    split_dir / "train_edges.csv",
    index=False
)
```

### Save Validation

```python
val_edges.to_csv(
    split_dir / "val_edges.csv",
    index=False
)
```

### Save Test

```python
test_edges.to_csv(
    split_dir / "test_edges.csv",
    index=False
)
```
