### Load CVE File

```python
import json
from pathlib import Path

project = Path(r"C:\Users\HP\Documents\python\bio informatics\final exam\ThreatGraphx")

cve_file = project / "data" / "raw" / "cve" / "nvdcve-2.0-recent.json"

with open(cve_file, "r", encoding="utf-8") as f:
    cve_data = json.load(f)
```

### Top-Level Keys

```python
cve_data.keys()
```

### Inspect One Record

```python
cve_data["vulnerabilities"][0]
```

### Extract One CVE ID

```python
cve_data["vulnerabilities"][0]["cve"]["id"]
```

### Extracting All CVE IDs

```python
cve_ids = []

for item in cve_data["vulnerabilities"]:
    cve_ids.append(item["cve"]["id"])

len(cve_ids)
```

### First 10 IDs

```python
cve_ids[:10]
```

###  convert to DataFrame 

```python
import pandas as pd

cve_df = pd.DataFrame({
    "cve_id": cve_ids
})

cve_df.head()
```

## Save processed Dataset

```python
output_file = project / "data" / "processed" / "cve_nodes.csv"

cve_df.to_csv(output_file, index=False)

print("saved")
```
