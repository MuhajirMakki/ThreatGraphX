### Imports

```python
import pandas as pd

from pathlib import Path

from pykeen.triples import TriplesFactory
```

### Project Path

```python
project = Path(
    r"C:\Users\HP\Documents\python\bio informatics\final exam\ThreatGraphx"
)
```

### Load Triple Files

```python
train_df = pd.read_csv(
    project / "data" / "triples" / "train_triples.csv"
)

val_df = pd.read_csv(
    project / "data" / "triples" / "val_triples.csv"
)

test_df = pd.read_csv(
    project / "data" / "triples" / "test_triples.csv"
)

print(train_df.shape)
print(val_df.shape)
print(test_df.shape)
```

### Convert To Strings
- PyKEEN expects strings.

### Convert To Strings

```python
train = train_df.astype(str)

val = val_df.astype(str)

test = test_df.astype(str)
```

### Create TriplesFactory

```python
train_tf = TriplesFactory.from_labeled_triples(
    train.values
)

val_tf = TriplesFactory.from_labeled_triples(
    val.values
)

test_tf = TriplesFactory.from_labeled_triples(
    test.values
)
```

```python
train_tf
```

----------------

## TransE MODEL

### Import Pipeline

```python
from pykeen.pipeline import pipeline
```

### Train TransE

```python
result = pipeline(
    
    training=train_tf,
    
    validation=val_tf,
    
    testing=test_tf,
    
    model="TransE",
    
    training_kwargs=dict(
        num_epochs=50
    ),
    
    model_kwargs=dict(
        embedding_dim=128
    ),
    
    random_seed=42
)
```

### Results

```python
result
```

--------------

### Evaluation Metrics

```python
metrics = result.metric_results.to_dict()

metrics
```

### Print Important Metrics

```python
print("MRR:",
      metrics["both"]["realistic"]["inverse_harmonic_mean_rank"])

print("Hits@1:",
      metrics["both"]["realistic"]["hits_at_1"])

print("Hits@3:",
      metrics["both"]["realistic"]["hits_at_3"])

print("Hits@10:",
      metrics["both"]["realistic"]["hits_at_10"])
```

### Save Model

```python
result.save_to_directory(
    model_dir / "TransE"
)
```
