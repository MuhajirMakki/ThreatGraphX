### BUILD SOFTWARE EDGES

```python
software_edges = []

for vuln in cve_data["vulnerabilities"]:

    cve = vuln["cve"]

    cve_id = cve["id"]

    if "affected" not in cve:
        continue

    for affected_block in cve["affected"]:

        if "affectedData" not in affected_block:
            continue

        for item in affected_block["affectedData"]:

            vendor = item.get("vendor", "").strip()
            product = item.get("product", "").strip()

            if product:

                software_name = f"{vendor}:{product}"

                software_edges.append([
                    cve_id,
                    "affects",
                    software_name
                ])
```

### CHECK COUNT

```python
len(software_edges)
```

### REMOVE DUPLICATES

```python
software_edges = list(
    set(
        tuple(x)
        for x in software_edges
    )
)

len(software_edges)
```

### Explore Data points

```python
software_edges[:5]
```

### Make dataframe

```python
import pandas as pd

software_df = pd.DataFrame(
    software_edges,
    columns=["source", "relation", "target"]
)

software_df.head()
```

### Save

```python
output_file = (
    project
    / "data"
    / "processed"
    / "software_edges.csv"
)

software_df.to_csv(
    output_file,
    index=False
)

print("saved")
```
