### imports

```python
import pandas as pd
import pickle

from pathlib import Path
```

### Project Path

```python
project = Path(
    r"C:\Users\HP\Documents\python\bio informatics\final exam\ThreatGraphx"
)
```

### Load Dictionaries

```python
with open(
    project / "data" / "ml_ready" / "node2id.pkl",
    "rb"
) as f:
    
    node2id = pickle.load(f)

with open(
    project / "data" / "ml_ready" / "relation2id.pkl",
    "rb"
) as f:
    
    relation2id = pickle.load(f)

print(len(node2id))
print(len(relation2id))
```

### Load Splits

```python
train_df = pd.read_csv(
    project / "data" / "splits" / "train_edges.csv"
)

val_df = pd.read_csv(
    project / "data" / "splits" / "val_edges.csv"
)

test_df = pd.read_csv(
    project / "data" / "splits" / "test_edges.csv"
)

print(train_df.shape)
print(val_df.shape)
print(test_df.shape)
```

### Conversion Function

```python
def convert_to_ids(df):

    triples = pd.DataFrame()

    triples["head"] = df["source"].map(node2id)

    triples["relation"] = df["relation"].map(relation2id)

    triples["tail"] = df["target"].map(node2id)

    return triples
```

### Convert All Splits

```python
train_triples = convert_to_ids(train_df)

val_triples = convert_to_ids(val_df)

test_triples = convert_to_ids(test_df)
```

### Check Missing Values

```python
print(train_triples.isna().sum())

print(val_triples.isna().sum())

print(test_triples.isna().sum())
```

### Save Triple Files

### Save Train

```python
train_triples.to_csv(
    triple_dir / "train_triples.csv",
    index=False
)
```

```python
test_triples.to_csv(
    triple_dir / "test_triples.csv",
    index=False
)
```
