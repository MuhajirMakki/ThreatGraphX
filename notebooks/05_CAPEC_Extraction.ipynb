### Load CAPEC XML

```python
import xml.etree.ElementTree as ET
from pathlib import Path

project = Path(r"C:\Users\HP\Documents\python\bio informatics\final exam\ThreatGraphx")

capec_file = project / "data" / "raw" / "capec" / "capec_latest.xml"

tree = ET.parse(capec_file)
root = tree.getroot()
```

```python
print(root.tag)
```

### Find first attack pattern

```python
for elem in root.iter():
    if elem.tag.endswith("Attack_Pattern"):
        print(elem.attrib)
        break
```

#### Extract All CAPEC IDs

```python
capec_ids = []

for elem in root.iter():
    
    if elem.tag.endswith("Attack_Pattern"):
        
        if "ID" in elem.attrib:
            
            capec_ids.append(
                "CAPEC-" + elem.attrib["ID"]
            )

len(capec_ids)
```

### Inspect First 20

```python
capec_ids[:20]
```

### making DataFrame

```python
import pandas as pd

capec_df = pd.DataFrame({
    "capec_id": capec_ids
})

capec_df.head()
```

### Remove Duplicates

```python
capec_df = capec_df.drop_duplicates()

len(capec_df)
```

### Save 

```python
output_file = project / "data" / "processed" / "capec_nodes.csv"

capec_df.to_csv(output_file, index=False)

print("saved")
```
